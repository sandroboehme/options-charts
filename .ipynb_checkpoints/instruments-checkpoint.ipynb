## All available options as of 29.4.23

Got them via

`curl "https://history.deribit.com/api/v2/public/get_instruments?currency=BTC&kind=option&include_old=true&expired=true" >instruments.json`

```python
import pandas as pd
import plotly.graph_objects as go
import datetime as datetime
```

## Expiration:

```python
expiration = "31DEC21"
```

```python
df = pd.read_json('instruments.json')

df = pd.json_normalize(df.result)
df["creation_date"] = pd.to_datetime(df["creation_timestamp"], unit='ms').dt.date
df["expiration_date"] = pd.to_datetime(df["expiration_timestamp"], unit='ms').dt.date
df['duration_days'] = round((df['expiration_timestamp'] - df['creation_timestamp']) / (1000*60*60*24), 0) - 2
df = df.loc[:, ['instrument_name', 'settlement_period', 'creation_date', 'expiration_date', 'duration_days']].copy()
#BTC-31MAR23-20000-C
df
```

```python
df2 = df.loc[df["instrument_name"].str.contains(expiration)]
#df2['strike'] = df2['instrument_name'].str.split('-').str[2]
#df2['strike'] = pd.to_numeric(df2['strike'])
df2 = df2[~df2['instrument_name'].str.endswith('P')] # drop Puts
df2 = df2[df2['duration_days'] >= 100] # copy only durations >= 200

with pd.option_context('display.max_rows', 100):
    display(df2)
```
